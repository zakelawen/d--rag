```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "0,1,2"

import torch
from tqdm import tqdm
import gc
import sys
import json
from datetime import datetime
from transformers import (
    AutoModelForCausalLM, 
    AutoTokenizer, 
    LogitsProcessorList, 
    BitsAndBytesConfig,
    StoppingCriteria,
    StoppingCriteriaList
)

from seper.uncertainty_measures.semantic_entropy import EntailmentDeberta
from seper.calculate import (
    gen_answers_batch,
    calculate_uncertainty_soft_batch,
    create_collate_fn,
    process_item_for_seper
)


model_path = ''

num_generations = 10
sub_batch_size = 10
temperature = 1.0
max_new_tokens = 20
max_context_words = 20000


computation_chunk_size = 8


FLUSH_EVERY = 50
CHECKPOINT_EVERY = 100
CLEAR_CACHE_EVERY = 1
prompt_type = 'default'



quantization_config = BitsAndBytesConfig(
    load_in_8bit=True
)


max_memory = {
    0: "0GiB",      # GPU 0 不分配给语言模型（留给 NLI）
    1: "22GiB",     # GPU 1 
    2: "22GiB",     # GPU 2
}


print("正在加载 tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(
    model_path, 
    token_type_ids=None, 
    clean_up_tokenization_spaces=False
)

print("正在加载语言模型（8-bit 量化，部署到 GPU 1,2）...")
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    quantization_config=quantization_config,
    device_map="auto",
    max_memory=max_memory,
    attn_implementation="flash_attention_2",
)
model.eval()
print(f"语言模型设备分布: {model.hf_device_map}")


class SimpleGenerator:
    def __init__(self, model, tokenizer, max_new_tokens=20):
        self.model = model
        self.tokenizer = tokenizer
        self.max_new_tokens = max_new_tokens
        self.model_name = model_path
        self.stop_sequences = ['\n\n\n\n', '\n\n\n', '\n\n', '\n', 'Question:', 'Context:', tokenizer.eos_token]
        self.token_limit = 128000
        self.logits_processor = LogitsProcessorList([])
    
    def batch_predict(self, input_data, num_generations, temperature, return_full=False, device='cuda', max_new_tokens=None):
        batch_input = [input_data] * num_generations
        inputs = self.tokenizer(batch_input, return_tensors="pt").to(device)
        
        if 'token_type_ids' in inputs:
            del inputs['token_type_ids']
        pad_token_id = self.tokenizer.eos_token_id
        
        with torch.no_grad():
            outputs = self.model.generate(
                **inputs,
                max_new_tokens=max_new_tokens if max_new_tokens is not None else self.max_new_tokens,
                return_dict_in_generate=True,
                output_scores=True,
                temperature=temperature,
                do_sample=True,
                pad_token_id=pad_token_id,
            )
        
        sliced_answers = []
        batch_n_generated = []
        
        for idx in range(len(outputs.sequences)):
            n_input_token = len(inputs['input_ids'][idx])
            answer = self.tokenizer.decode(outputs.sequences[idx][n_input_token:], skip_special_tokens=False)
            
            # 移除 stop sequences
            for stop in self.stop_sequences:
                if stop and answer.endswith(stop):
                    answer = answer[:-len(stop)]
            
            sliced_answers.append(answer.strip())
            n_generated = len(self.tokenizer.encode(answer.strip())) or 1
            batch_n_generated.append(n_generated)
        
        # 计算 log likelihoods
        transition_scores = self.model.compute_transition_scores(
            outputs.sequences, outputs.scores, normalize_logits=True
        )
        
        batch_log_likelihoods = []
        for idx, ts in enumerate(transition_scores):
            n_generated = batch_n_generated[idx]
            log_likelihoods = [score.item() for score in ts][:n_generated]
            if len(log_likelihoods) == 0:
                log_likelihoods = [0.0]
            batch_log_likelihoods.append(log_likelihoods)
        
        return sliced_answers, batch_log_likelihoods, None


generator = SimpleGenerator(model, tokenizer, max_new_tokens=max_new_tokens)

print("正在加载 NLI 蕴含模型（部署到 GPU 0）...")
device_nli = 'cuda:0'
entailment_model = EntailmentDeberta(device=device_nli)
entailment_model.model.eval()
print(f"NLI 模型已加载到: {device_nli}")


processed_dataset_path = ''
my_custom_dataset = []
try:
    with open(processed_dataset_path, 'r', encoding='utf-8') as f:
        for line in f:
            if line.strip():
                my_custom_dataset.append(json.loads(line))
    print(f"成功加载数据集: {processed_dataset_path} （共 {len(my_custom_dataset)} 条）")
except FileNotFoundError:
    print(f"错误：找不到数据集文件 -> {processed_dataset_path}")
    sys.exit(1)

# 数据整理函数
keys = ['question', 'response_text', 'answers', 'likelihood', 'context_label', 'log_liks_agg', 'context']
seper_collate_fn = create_collate_fn(keys)

# 创建结果数据集的副本
result_dataset = [item.copy() for item in my_custom_dataset]


output_dir = ''
os.makedirs(output_dir, exist_ok=True)
current_time = datetime.now().strftime("%Y%m%d_%H%M%S")
output_path = os.path.join(output_dir, f'medexqa_llama_seper_results_{current_time}.jsonl')
checkpoint_path = os.path.join(output_dir, f'medexqa_llama_checkpoint_{current_time}.json')


start_idx = 0
processed_ids = set()
if os.path.exists(checkpoint_path):
    try:
        with open(checkpoint_path, 'r', encoding='utf-8') as f:
            checkpoint = json.load(f)
            start_idx = checkpoint.get('last_processed_idx', 0) + 1
            processed_ids = set(checkpoint.get('processed_ids', []))
        print(f"发现检查点，从索引 {start_idx} 继续处理")
    except Exception as e:
        print(f"读取检查点失败: {e}，从头开始")
        start_idx = 0
        processed_ids = set()


buffer = []
pbar = tqdm(
    range(start_idx, len(result_dataset)),
    initial=start_idx,
    total=len(result_dataset),
    desc="处理数据集",
    mininterval=1.0
)

for idx in pbar:
    item = result_dataset[idx]
    item_id = item.get('id', idx)

    if item_id in processed_ids:
        continue

    question = item.get('question', '')
    ground_truths = item.get('ground_truths', [''])
    docs = item.get('docs', [])

    # 基线：无上下文
    example_baseline = {'question': question, 'context': '', 'answers': ground_truths}
    result_baseline = gen_answers_batch(
        example_baseline, generator, temperature, num_generations,
        sub_batch_size, max_new_tokens, prompt_type, 'cuda', max_context_words
    )

    # 完整上下文
    full_context = '\n'.join([doc.get('text', '') for doc in docs]) if docs else ''
    example_full = {'question': question, 'context': full_context, 'answers': ground_truths}
    result_full = gen_answers_batch(
        example_full, generator, temperature, num_generations,
        sub_batch_size, max_new_tokens, prompt_type, 'cuda', max_context_words
    )

    # 部分上下文
    results_pieces = []
    if docs:
        for doc in docs:
            piece_context = doc.get('text', '')
            example_piece = {'question': question, 'context': piece_context, 'answers': ground_truths}
            rp = gen_answers_batch(
                example_piece, generator, temperature, num_generations,
                sub_batch_size, max_new_tokens, prompt_type, 'cuda', max_context_words
            )
            results_pieces.append(rp)

    # 计算 SePer
    with torch.no_grad():
        rb = process_item_for_seper(result_baseline)
        seper_baseline = calculate_uncertainty_soft_batch(
            seper_collate_fn([rb]), entailment_model, computation_chunk_size
        )[0]

        r_full = process_item_for_seper(result_full)
        seper_full = calculate_uncertainty_soft_batch(
            seper_collate_fn([r_full]), entailment_model, computation_chunk_size
        )[0]
        d_seper_full = seper_full - seper_baseline

        if results_pieces:
            r_pieces = [process_item_for_seper(rp) for rp in results_pieces]
            pieces_seper_list = calculate_uncertainty_soft_batch(
                seper_collate_fn(r_pieces), entailment_model, computation_chunk_size
            )
            for doc, seper_piece in zip(docs, pieces_seper_list):
                d_seper_piece = seper_piece - seper_baseline
                doc['seper'] = float(seper_piece)
                doc['d_seper'] = float(d_seper_piece)

    item['seper_baseline'] = float(seper_baseline)
    item['seper_full'] = float(seper_full)
    item['d_seper_full'] = float(d_seper_full)

    buffer.append(json.dumps(item, ensure_ascii=False))
    processed_ids.add(item_id)

    # 每条回收显存
    if (idx + 1) % CLEAR_CACHE_EVERY == 0:
        del result_baseline, result_full, results_pieces
        del rb, r_full, seper_baseline, seper_full, d_seper_full
        torch.cuda.empty_cache()
        gc.collect()

    # 定期写文件
    if len(buffer) >= FLUSH_EVERY:
        with open(output_path, 'a', encoding='utf-8') as f:
            f.write('\n'.join(buffer) + '\n')
        buffer.clear()

    # 定期写检查点
    if (idx + 1) % CHECKPOINT_EVERY == 0:
        checkpoint_data = {
            'last_processed_idx': idx,
            'processed_ids': list(processed_ids),
            'output_path': output_path
        }
        with open(checkpoint_path, 'w', encoding='utf-8') as f:
            json.dump(checkpoint_data, f)

# 写剩余缓冲
if buffer:
    with open(output_path, 'a', encoding='utf-8') as f:
        f.write('\n'.join(buffer) + '\n')
    buffer.clear()

# 最终检查点
checkpoint_data = {
    'last_processed_idx': len(result_dataset) - 1,
    'processed_ids': list(processed_ids),
    'output_path': output_path
}
with open(checkpoint_path, 'w', encoding='utf-8') as f:
    json.dump(checkpoint_data, f)

# 清理
del generator, entailment_model, model
torch.cuda.empty_cache()
gc.collect()
print(f"处理完成，结果已保存到: {output_path}")
```

```python


import os
import json
import time
from typing import List, Dict, Any, Optional

import torch
from transformers import AutoTokenizer, GenerationConfig
from gen_wrapper import LMWrapper
from tqdm import tqdm

import os
os.environ["CUDA_VISIBLE_DEVICES"] = "1,2"

# ====== 配置 ======
MODEL_NAME = ""
CACHE_DIR = "./cache"
OFFLOAD_DIR = "./offload"
DEVICE_MAP = "auto"
MAX_MEMORY = None

INPUT_JSONL = ""

HARMFUL_THRESHOLD = -0.0458 #adaptive 
USEFUL_THRESHOLD  =  0.1622  #adaptive 

MAX_NEW_TOKENS = 20
DO_SAMPLE = False  # 强制贪婪
GEN_BUDGET_TOKENS = 256
EVIDENCE_SEP = "\n\n---\n\n"

# 组合 α 列表
ALPHA_NEG_VALUES = [0.05, 0.1, 0.2]
ALPHA_POS_VALUES = [0.1, 0.5, 1.0]

# 输出前缀
BASE_OUTPUT_PATH = ""

# ====== 工具函数 ======
def load_jsonl(path: str) -> List[Dict[str, Any]]:
    data = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line: continue
            data.append(json.loads(line))
    return data

def build_stem(item: Dict[str, Any]) -> str:
    rq = item.get("raw_question")
    if rq and isinstance(rq, str) and rq.strip():
        stem = rq.strip()
        if not stem.lower().strip().endswith("answer:"):
            stem += "\nAnswer:"
        return stem
    q = (item.get("question") or "").strip()
    if not q:
        q = "Answer the multiple-choice question."
    if not q.lower().strip().endswith("answer:"):
        q += "\nAnswer:"
    return q

def token_len(tokenizer: AutoTokenizer, text: str) -> int:
    return tokenizer(text, return_tensors="pt", add_special_tokens=False).input_ids.size(1)

def build_pos_assistant_text(docs: List[Dict[str, Any]]) -> str:
    if not docs: return "No retrieved documents were provided."
    chunks = [f"[{i+1}] Title: {d.get('title') or f'doc_{i}'}\n{d.get('text') or ''}" for i, d in enumerate(docs)]
    body = EVIDENCE_SEP.join(chunks)
    prefix = ("The following retrieved documents are provided verbatim before the question. "
              "Answer the upcoming multiple-choice question with only the final option letter "
              "(e.g., A, B, C, D, or E) without explanations.")
    return prefix + "\n\n" + body

def build_neg_assistant_text(docs: List[Dict[str, Any]]) -> str:
    if not docs: return "No negative documents were provided."
    chunks = [f"[NEG-{i+1}] Title: {d.get('title') or f'doc_{i}'}\n{d.get('text') or ''}" for i, d in enumerate(docs)]
    body = EVIDENCE_SEP.join(chunks)
    prefix = ("The following retrieved documents are provided verbatim before the question. "
              "Answer the upcoming multiple-choice question with only the final option letter "
              "(e.g., A, B, C, D, or E) without explanations.")
    return prefix + "\n\n" + body

def apply_chat(tokenizer: AutoTokenizer, messages: List[Dict[str, str]]) -> str:
    return tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

def ensure_dir_for_file(path: str):
    os.makedirs(os.path.dirname(path), exist_ok=True)

def make_gen_config(tokenizer: AutoTokenizer,
                    do_sample: bool,
                    acd_alpha_pos: Optional[float]=None,
                    acd_alpha_neg: Optional[float]=None) -> GenerationConfig:
    kwargs = dict(
        max_new_tokens=MAX_NEW_TOKENS,
        pad_token_id=tokenizer.pad_token_id,
        eos_token_id=tokenizer.eos_token_id,
        do_sample=do_sample,  # False: 贪婪
        return_dict_in_generate=True,
        output_scores=False,
    )
    if acd_alpha_pos is not None:
        kwargs["acd_alpha_pos"] = float(acd_alpha_pos)
    if acd_alpha_neg is not None:
        kwargs["acd_alpha_neg"] = float(acd_alpha_neg)
    return GenerationConfig(**kwargs)

def choose_strategy(needs_retrieval: int, dscore: Optional[float]) -> str:
    if not needs_retrieval:
        return "noctx-normal"
    if dscore is None:
        return "ctx-normal"
    if dscore < HARMFUL_THRESHOLD:
        return "ctx-neg-acd"
    if dscore > USEFUL_THRESHOLD:
        return "ctx-pos-acd"
    return "ctx-normal"

def build_main_prompt(tokenizer: AutoTokenizer, stem: str) -> str:
    main_messages = [
        {"role": "system", "content": "You are a helpful assistant. Your answer must be only the letter of the correct option (e.g., A, B, C, D, or E)."},
        {"role": "user", "content": stem},
    ]
    return apply_chat(tokenizer, main_messages)

def build_ctx_prompt(tokenizer: AutoTokenizer, stem: str, docs: List[Dict[str, Any]], polarity: str) -> (str, str):
    if polarity == "pos":
        ctx_text = build_pos_assistant_text(docs)
    else:
        ctx_text = build_neg_assistant_text(docs)
    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "assistant", "content": ctx_text},
        {"role": "user", "content": stem},
    ]
    return apply_chat(tokenizer, messages), ctx_text[:1200]


def run_single_combo(alpha_pos: float, alpha_neg: float,
                     base_output_path: str,
                     dataset: List[Dict[str, Any]],
                     tokenizer: AutoTokenizer,
                     model: LMWrapper):
    out_path = f"{base_output_path}_llama_pos_{alpha_pos}_neg_{alpha_neg}.jsonl"
    ensure_dir_for_file(out_path)
    print(f"\n--- Running combo: POS={alpha_pos} | NEG={alpha_neg} ---")
    print(f"Output -> {out_path}")

    with open(out_path, "w", encoding="utf-8") as out_f:
        pbar = tqdm(enumerate(dataset), total=len(dataset), desc=f"pos={alpha_pos},neg={alpha_neg}", unit="sample")
        t0 = time.time()
        for idx, item in pbar:
            stem = build_stem(item)
            docs = item.get("docs") or []
            needs_retrieval = int(item.get("needs_retrieval", 1))
            dscore = item.get("d_seper_full", None)

            # 构主 prompt
            prompt_main = build_main_prompt(tokenizer, stem)

            def _check_len(text: str, tag: str):
                L = token_len(tokenizer, text)
                model_max = getattr(tokenizer, "model_max_length", 32768)
                hard_limit = max(512, model_max - GEN_BUDGET_TOKENS)
                if L > hard_limit:
                    pbar.write(f"[Warn] #{idx} {tag} tokens={L} > {hard_limit} (model≈{model_max}).")
                return L
            main_len = _check_len(prompt_main, "main")

            strategy = choose_strategy(needs_retrieval, dscore)

            # 公共元信息
            meta = {
                "index": idx,
                "question": item.get("question"),
                "raw_question": item.get("raw_question"),
                "ground_truths": item.get("ground_truths"),
                "needs_retrieval": needs_retrieval,
                "d_seper_full_score": dscore,
                "doc_titles": [d.get("title") for d in docs] if docs else [],
                # 记录当前组合（即使该样本策略没用到某个 α）
                "combo_alpha_pos": alpha_pos,
                "combo_alpha_neg": alpha_neg,
                "strategy": strategy,
                "main_token_len": main_len,
            }

            # 执行
            if strategy == "noctx-normal":
                inputs = tokenizer(prompt_main, return_tensors="pt").to(model.device).input_ids
                gen_cfg = make_gen_config(tokenizer, DO_SAMPLE)
                with torch.no_grad():
                    out = model.generate_contrast(inputs=inputs, generation_config=gen_cfg, mode=None)
                gen_text = tokenizer.decode(out.sequences[0][inputs.shape[1]:], skip_special_tokens=True).strip()
                rec = dict(meta, decoding_strategy="noctx-normal",
                           acd_alpha_pos=None, acd_alpha_neg=None,
                           model_output_raw=gen_text)
                out_f.write(json.dumps(rec, ensure_ascii=False) + "\n")

            elif strategy == "ctx-normal":
                prompt_ctx, evidence_head = build_ctx_prompt(tokenizer, stem, docs, "pos")  # 中性前缀=pos版本
                _ = _check_len(prompt_ctx, "ctx-normal")
                inputs = tokenizer(prompt_ctx, return_tensors="pt").to(model.device).input_ids
                gen_cfg = make_gen_config(tokenizer, DO_SAMPLE)
                with torch.no_grad():
                    out = model.generate_contrast(inputs=inputs, generation_config=gen_cfg, mode=None)
                gen_text = tokenizer.decode(out.sequences[0][inputs.shape[1]:], skip_special_tokens=True).strip()
                rec = dict(meta, decoding_strategy="ctx-normal",
                           acd_alpha_pos=None, acd_alpha_neg=None,
                           model_output_raw=gen_text,
                           strategy_details={"type":"ctx-normal","token_len": token_len(tokenizer, prompt_ctx),"evidence_head": evidence_head})
                out_f.write(json.dumps(rec, ensure_ascii=False) + "\n")

            elif strategy == "ctx-neg-acd":
                prompt_neg, evidence_head = build_ctx_prompt(tokenizer, stem, docs, "neg")
                batch = tokenizer([prompt_main, prompt_neg], return_tensors="pt", padding="longest").to(model.device)
                inputs_main, inputs_neg = batch.input_ids[0:1], batch.input_ids[1:2]
                input_len = inputs_main.shape[1]
                gen_cfg = make_gen_config(tokenizer, DO_SAMPLE, acd_alpha_neg=alpha_neg)
                with torch.no_grad():
                    out = model.generate_contrast(
                        inputs=inputs_main,
                        inputs_neg=inputs_neg,
                        generation_config=gen_cfg,
                        mode="acd",
                    )
                gen_text = tokenizer.decode(out.sequences[0][input_len:], skip_special_tokens=True).strip()
                rec = dict(meta, decoding_strategy="ctx-neg-acd",
                           acd_alpha_pos=None, acd_alpha_neg=alpha_neg,
                           model_output_raw=gen_text,
                           strategy_details={"type":"ctx-neg-acd","token_len": token_len(tokenizer, prompt_neg),"evidence_head": evidence_head})
                out_f.write(json.dumps(rec, ensure_ascii=False) + "\n")

            elif strategy == "ctx-pos-acd":
                prompt_pos, evidence_head = build_ctx_prompt(tokenizer, stem, docs, "pos")
                batch = tokenizer([prompt_main, prompt_pos], return_tensors="pt", padding="longest").to(model.device)
                inputs_main, inputs_pos = batch.input_ids[0:1], batch.input_ids[1:2]
                input_len = inputs_main.shape[1]
                gen_cfg = make_gen_config(tokenizer, DO_SAMPLE, acd_alpha_pos=alpha_pos)
                with torch.no_grad():
                    out = model.generate_contrast(
                        inputs=inputs_main,
                        inputs_pos=inputs_pos,
                        generation_config=gen_cfg,
                        mode="acd",
                    )
                gen_text = tokenizer.decode(out.sequences[0][input_len:], skip_special_tokens=True).strip()
                rec = dict(meta, decoding_strategy="ctx-pos-acd",
                           acd_alpha_pos=alpha_pos, acd_alpha_neg=None,
                           model_output_raw=gen_text,
                           strategy_details={"type":"ctx-pos-acd","token_len": token_len(tokenizer, prompt_pos),"evidence_head": evidence_head})
                out_f.write(json.dumps(rec, ensure_ascii=False) + "\n")

            if torch.cuda.is_available():
                torch.cuda.empty_cache()

        pbar.close()
    print(f"Done combo pos={alpha_pos}, neg={alpha_neg} -> {out_path}")


def main():
    assert HARMFUL_THRESHOLD < USEFUL_THRESHOLD, "要求：HARMFUL_THRESHOLD < USEFUL_THRESHOLD"

    print("Loading model and tokenizer...")
    model = LMWrapper(
        model_name_or_path=MODEL_NAME,
        cache_dir=CACHE_DIR,
        device_map=DEVICE_MAP,
        max_memory=MAX_MEMORY,
        offload_folder=OFFLOAD_DIR
    )
    tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token
    tokenizer.padding_side = "left"
    print("Ready.")
    print("-" * 30)

    if not os.path.exists(INPUT_JSONL):
        print(f"Error: dataset file not found -> {INPUT_JSONL}")
        return
    dataset = load_jsonl(INPUT_JSONL)
    print(f"Loaded {len(dataset)} items.")
    print("-" * 30)

    for alpha_pos in ALPHA_POS_VALUES:
        for alpha_neg in ALPHA_NEG_VALUES:
            run_single_combo(alpha_pos, alpha_neg, BASE_OUTPUT_PATH, dataset, tokenizer, model)

if __name__ == "__main__":
    main()
```
