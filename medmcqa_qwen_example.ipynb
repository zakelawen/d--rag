```python
# 为了解决 'ndarray is not JSON serializable' 的错误，
# 我们需要定义一个自定义的转换函数，告诉json库如何处理这些特殊类型。

import json
import os
import numpy as np

# 尝试导入torch，因为数据中也可能包含PyTorch张量，但如果未安装也不报错
try:
    import torch
    HAS_TORCH = True
except ImportError:
    HAS_TORCH = False

def custom_serializer(obj):
    """
    自定义JSON序列化函数，用于处理默认库无法处理的对象。
    此函数将NumPy数值类型、数组和PyTorch张量转换为Python原生类型。
    """
    # 检查是否为NumPy的数值类型 (e.g., np.float32, np.int64)
    # 并将其转换为对应的Python原生类型 (e.g., float, int)
    if isinstance(obj, np.number):
        return obj.item()
    
    # 如果对象是NumPy数组，则调用 .tolist() 方法转换
    if isinstance(obj, np.ndarray):
        return obj.tolist()
    
    # 如果torch已导入且对象是PyTorch张量，也调用 .tolist()
    if HAS_TORCH and isinstance(obj, torch.Tensor):
        return obj.tolist()
    
    # 对于其他任何json无法处理的类型，抛出原始的TypeError
    raise TypeError(f"Object of type {type(obj).__name__} is not JSON serializable")


# 定义输出文件的名称
output_filename = 'labeled_results_qwen_1.json'

try:
    with open(output_filename, 'w', encoding='utf-8') as f:
        # 使用 json.dump 保存，并通过 'default' 参数传入我们的自定义函数
        # indent=4 让JSON文件格式化，更易于阅读
        # ensure_ascii=False 确保中文等非ASCII字符能正确显示
        json.dump(results, f, ensure_ascii=False, indent=4, default=custom_serializer)
        
    print(f"✅ 成功！已将带标签的结果保存到文件: {os.path.abspath(output_filename)}")

except Exception as e:
    print(f"❌ 保存文件时出错: {e}")

# 打印第一条结果，检查 'label' 是否已成功添加
if results:
    print("\n第一条结果预览:")
    # 在打印预览时，同样需要使用我们的自定义序列化函数
    print(json.dumps(results[0], indent=4, ensure_ascii=False, default=custom_serializer))
```

```python
# -*- coding: utf-8 -*-
"""
使用 LR 和 MLP 模型生成真值分数，并使用标准化的评估框架进行性能比较。

该脚本执行以下步骤：
1. 加载包含特征和标签的数据。
2. 将数据分割为训练集和测试集，并保存测试集的原始索引。
3. 训练一个逻辑回归（LR）模型和多个多层感知机（MLP）模型（不同隐藏层组合）。
4. 使用训练好的模型在测试集上生成置信度分数，这些分数被视为“真值分数”。
5. 调用一个统一的评估函数（metric_score），使用PRR、AUROC等指标，
   来评估这些模型生成的真值分数的质量。
"""

# 1. 导入
import os
import json
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_auc_score, precision_recall_curve, auc

# ==============================================================================
# 2. 评估函数
# ==============================================================================

def area_under_accuracy_coverage_curve(t_s, acc):
    """计算准确率-覆盖率曲线下的面积 (AUACC)。"""
    df = pd.DataFrame({"t_s": t_s, "acc": acc}).sort_values("t_s", ascending=False)
    df["acc_mean"] = df["acc"].expanding().mean()
    return auc(np.linspace(0, 1, len(df)), df["acc_mean"])

def normalize(target):
    """将一个数组的值归一化到 [0, 1] 区间。"""
    min_t, max_t = np.min(target), np.max(target)
    if np.isclose(min_t, max_t):
        min_t -= 1
        max_t += 1
    target = (np.array(target) - min_t) / (max_t - min_t)
    return target

def prediction_rejection_curve(estimator, target):
    """计算预测拒绝曲线的原始分数（未归一化）。"""
    target = normalize(target)  # 分数越高代表越正确
    ue = np.array(estimator)
    num_obs = len(ue)
    # 按评估器分数降序排序（置信度最高的在前）
    ue_argsort = np.argsort(ue)[::-1]
    sorted_metrics = np.array(target)[ue_argsort]
    cumsum = np.cumsum(sorted_metrics)
    scores = (cumsum / np.arange(1, num_obs + 1))
    prr_score = np.sum(scores) / num_obs
    return prr_score

def get_random_scores(function, metrics, num_iter=1000, seed=42):
    """为给定的指标函数计算随机基线分数。"""
    np.random.seed(seed)
    rand_scores_template = np.arange(len(metrics))
    value = []
    for _ in range(num_iter):
        np.random.shuffle(rand_scores_template)
        rand_val = function(rand_scores_template, metrics)
        value.append(rand_val)
    return np.mean(value)

def metric_score(
    metric_names: list[str],
    generation_correctness: list,
    truth_values: list[float],
    seed: int = 0,
) -> dict:
    """
    为真值分数估计计算多种评估指标。
    这是主要的评估工具函数。
    """
    eval_dict = {}
    generation_correctness = np.array(generation_correctness)
    truth_values = np.array(truth_values)
    
    # --- PRR 计算 ---
    if "prr" in metric_names:
        # 1. 模型的原始分
        ue_prr = prediction_rejection_curve(truth_values, generation_correctness)
        # 2. 理想情况（神谕）的原始分
        orc_prr = prediction_rejection_curve(generation_correctness, generation_correctness)
        # 3. 随机情况的原始分（稳健的基线）
        rand_prr = get_random_scores(prediction_rejection_curve, generation_correctness, seed=seed)

        # 归一化得到最终的PRR分数
        if not np.isclose(orc_prr, rand_prr):
            prr_final = (ue_prr - rand_prr) / (orc_prr - rand_prr)
        else:
            prr_final = 0.0  # 避免除以零
        eval_dict["prr"] = prr_final

    # --- 其他指标 ---
    if "auroc" in metric_names:
        try:
            eval_dict["auroc"] = roc_auc_score(generation_correctness, truth_values)
        except ValueError:
            eval_dict["auroc"] = 0.5

    if "auprc" in metric_names:
        precision, recall, _ = precision_recall_curve(generation_correctness, truth_values)
        eval_dict["auprc"] = auc(recall, precision)
        
    if "auarc" in metric_names:
        eval_dict["auarc"] = area_under_accuracy_coverage_curve(truth_values, generation_correctness)
        
    return eval_dict

# ==============================================================================
# 3. 主执行流程
# ==============================================================================

def main():
    """
    主函数，运行整个实验。
    """
    # --- 配置 ---
    DATA_FILE_PATH = '/home/qluai/zjs/TruthTorchLM-main/medmcqa/other_llm_qwen/final_simplified_results_qwen_1.json' # <--- 请修改为您的数据文件路径
    MLP_HIDDEN_LAYERS_LIST = [(32,), (64,), (32, 32), (32, 64), (32, 64, 128)]  # MLP 模型的隐藏层结构列表
    METRICS_TO_CALCULATE = ["prr", "auroc", "auprc", "auarc"] # 您希望计算的指标列表

    # --- 步骤 1: 加载并分割数据 ---
    print("正在加载数据...")
    with open(DATA_FILE_PATH, 'r', encoding='utf-8') as f:
        all_data = json.load(f)

    # 假设每个item都有'normalized_truth_values'作为特征, 'label'作为正确性标签
    try:
        X = np.array([item['normalized_truth_values'] for item in all_data])
        y = np.array([item['label'] for item in all_data])
        indices = np.arange(len(X)) # 创建原始索引
    except KeyError:
        print("错误: 数据文件必须为每个项目包含 'normalized_truth_values' 和 'label'。")
        return

    # 分割数据时，同时分割原始索引
    X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
        X, y, indices, test_size=0.3, random_state=42, stratify=y
    )
    print(f"数据已分割为 {len(X_train)} 个训练样本和 {len(X_test)} 个测试样本。\n")

    # 保存测试集对应的原始索引
    with open("test_indices_mixtral_1.json", "w") as f:
        json.dump(idx_test.tolist(), f)
    absolute_path = os.path.abspath("test_indices_qwen_1.json")
    print(f"测试集对应的原始索引已保存到: {absolute_path}")

    # --- 步骤 2: 训练模型 ---
    print("--- 正在训练模型 ---")
    
    # 训练逻辑回归模型
    print("正在训练逻辑回归模型...")
    lr_model = LogisticRegression(max_iter=1000, random_state=42)
    lr_model.fit(X_train, y_train)
    print("逻辑回归模型训练完成。\n")

    # 训练多个 MLP 模型
    mlp_models = {}
    for hidden_layers in MLP_HIDDEN_LAYERS_LIST:
        print(f"正在训练MLP模型，结构为 {hidden_layers}...")
        mlp_model = MLPClassifier(hidden_layer_sizes=hidden_layers, max_iter=1000, random_state=42)
        mlp_model.fit(X_train, y_train)
        mlp_models[hidden_layers] = mlp_model
        print(f"MLP模型（{hidden_layers}）训练完成。\n")

    # --- 步骤 3: 在测试集上生成真值分数 ---
    print("--- 正在生成真值分数 ---")
    truth_values_lr = lr_model.predict_proba(X_test)[:, 1]
    truth_values_mlp = {hidden_layers: model.predict_proba(X_test)[:, 1] for hidden_layers, model in mlp_models.items()}
    print("真值分数生成完毕。\n")

    # --- 步骤 4: 评估生成的真值分数 ---
    print("--- 正在评估真值分数质量 ---")
    
    # 评估逻辑回归的分数
    lr_metrics = metric_score(
        metric_names=METRICS_TO_CALCULATE,
        generation_correctness=y_test,
        truth_values=truth_values_lr,
        seed=42
    )
    
    # 评估多个 MLP 的分数
    mlp_metrics = {}
    for hidden_layers in MLP_HIDDEN_LAYERS_LIST:
        mlp_metrics[hidden_layers] = metric_score(
            metric_names=METRICS_TO_CALCULATE,
            generation_correctness=y_test,
            truth_values=truth_values_mlp[hidden_layers],
            seed=42
        )

    # --- 步骤 5: 展示结果 ---
    print("\n--- 评估结果 ---\n")
    print(f"{'指标':<10} | {'逻辑回归':<20} | " + " | ".join([f"MLP {hl}" for hl in MLP_HIDDEN_LAYERS_LIST]))
    print("-" * (55 + 22 * len(MLP_HIDDEN_LAYERS_LIST)))
    for metric_name in METRICS_TO_CALCULATE:
        lr_val = lr_metrics.get(metric_name, 'N/A')
        mlp_vals = [mlp_metrics[hl].get(metric_name, 'N/A') for hl in MLP_HIDDEN_LAYERS_LIST]
        print(f"{metric_name.upper():<10} | {lr_val:<20.4f} | " + " | ".join([f"{val:<20.4f}" for val in mlp_vals]))

if __name__ == "__main__":
    main()
```

```python
# 为了解决 'ndarray is not JSON serializable' 的错误，
# 我们需要定义一个自定义的转换函数，告诉json库如何处理这些特殊类型。

import json
import os
import numpy as np

# 尝试导入torch，因为数据中也可能包含PyTorch张量，但如果未安装也不报错
try:
    import torch
    HAS_TORCH = True
except ImportError:
    HAS_TORCH = False

def custom_serializer(obj):
    """
    自定义JSON序列化函数，用于处理默认库无法处理的对象。
    此函数将NumPy数值类型、数组和PyTorch张量转换为Python原生类型。
    """
    # 检查是否为NumPy的数值类型 (e.g., np.float32, np.int64)
    # 并将其转换为对应的Python原生类型 (e.g., float, int)
    if isinstance(obj, np.number):
        return obj.item()
    
    # 如果对象是NumPy数组，则调用 .tolist() 方法转换
    if isinstance(obj, np.ndarray):
        return obj.tolist()
    
    # 如果torch已导入且对象是PyTorch张量，也调用 .tolist()
    if HAS_TORCH and isinstance(obj, torch.Tensor):
        return obj.tolist()
    
    # 对于其他任何json无法处理的类型，抛出原始的TypeError
    raise TypeError(f"Object of type {type(obj).__name__} is not JSON serializable")


# 定义输出文件的名称
output_filename = 'labeled_results_for_calibrating_qwen.json'

try:
    with open(output_filename, 'w', encoding='utf-8') as f:
        # 使用 json.dump 保存，并通过 'default' 参数传入我们的自定义函数
        # indent=4 让JSON文件格式化，更易于阅读
        # ensure_ascii=False 确保中文等非ASCII字符能正确显示
        json.dump(results, f, ensure_ascii=False, indent=4, default=custom_serializer)
        
    print(f"✅ 成功！已将带标签的结果保存到文件: {os.path.abspath(output_filename)}")

except Exception as e:
    print(f"❌ 保存文件时出错: {e}")

# 打印第一条结果，检查 'label' 是否已成功添加
if results:
    print("\n第一条结果预览:")
    # 在打印预览时，同样需要使用我们的自定义序列化函数
    print(json.dumps(results[0], indent=4, ensure_ascii=False, default=custom_serializer))
```

```python
# -*- coding: utf-8 -*-
"""
集成等渗回归校准功能，对 LR 和 MLP 模型生成的真值分数进行校准和评估。
新增功能：保存测试集样本对应的原始数据索引。

该脚本执行以下步骤：
1. 加载主数据集，并分割为训练集和测试集。
2. 加载独立的校准数据集。
3. 在训练集上训练 LR 和 MLP 模型。
4. 在校准集上，为 LR 和 MLP 模型分别拟合一个 IsotonicRegression 校准器。
5. 在测试集上，生成未校准和校准后的分数，并保存测试集索引。
6. 全面评估和对比未校准与校准后分数的质量。
"""

# 1. Imports
import json
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_auc_score, precision_recall_curve, auc
import sklearn.isotonic as isotonic

# ==============================================================================
# 2. 校准器与评估函数
# ==============================================================================

class Normalizer:
    def fit(self, *args, **kwargs):
        pass
    def __call__(self, *args, **kwargs):
        pass

class IsotonicRegression(Normalizer):
    """
    使用 sklearn.isotonic.IsotonicRegression 的校准器封装类。
    """
    def __init__(self, y_min: float = 0.0, y_max: float = 1.0):
        self.iso_reg = isotonic.IsotonicRegression(
            y_min=y_min, y_max=y_max, out_of_bounds="clip"
        )

    def fit(self, generation_performance_scores: list, truth_values: list):
        """使用未校准的分数和真实标签来拟合校准器。"""
        self.iso_reg.fit(truth_values, generation_performance_scores)

    def __call__(self, truth_value):
        """对单个真值分数进行校准。"""
        return self.iso_reg.predict([truth_value])[0]

def area_under_accuracy_coverage_curve(t_s, acc):
    df = pd.DataFrame({"t_s": t_s, "acc": acc}).sort_values("t_s", ascending=False)
    df["acc_mean"] = df["acc"].expanding().mean()
    return auc(np.linspace(0, 1, len(df)), df["acc_mean"])

def normalize(target):
    min_t, max_t = np.min(target), np.max(target)
    if np.isclose(min_t, max_t):
        min_t -= 1
        max_t += 1
    return (np.array(target) - min_t) / (max_t - min_t)

def prediction_rejection_curve(estimator, target):
    target = normalize(target)
    ue = np.array(estimator)
    num_obs = len(ue)
    ue_argsort = np.argsort(ue)[::-1]
    sorted_metrics = np.array(target)[ue_argsort]
    cumsum = np.cumsum(sorted_metrics)
    scores = (cumsum / np.arange(1, num_obs + 1))
    return np.sum(scores) / num_obs

def get_random_scores(function, metrics, num_iter=1000, seed=42):
    np.random.seed(seed)
    rand_scores_template = np.arange(len(metrics))
    value = []
    for _ in range(num_iter):
        np.random.shuffle(rand_scores_template)
        rand_val = function(rand_scores_template, metrics)
        value.append(rand_val)
    return np.mean(value)

def metric_score(metric_names, generation_correctness, truth_values, seed=0):
    eval_dict = {}
    generation_correctness = np.array(generation_correctness)
    truth_values = np.array(truth_values)
    
    if "prr" in metric_names:
        ue_prr = prediction_rejection_curve(truth_values, generation_correctness)
        orc_prr = prediction_rejection_curve(generation_correctness, generation_correctness)
        rand_prr = get_random_scores(prediction_rejection_curve, generation_correctness, seed=seed)
        if not np.isclose(orc_prr, rand_prr):
            eval_dict["prr"] = (ue_prr - rand_prr) / (orc_prr - rand_prr)
        else:
            eval_dict["prr"] = 0.0
            
    if "auroc" in metric_names:
        try:
            eval_dict["auroc"] = roc_auc_score(generation_correctness, truth_values)
        except ValueError:
            eval_dict["auroc"] = 0.5

    if "auprc" in metric_names:
        precision, recall, _ = precision_recall_curve(generation_correctness, truth_values)
        eval_dict["auprc"] = auc(recall, precision)
        
    if "auarc" in metric_names:
        eval_dict["auarc"] = area_under_accuracy_coverage_curve(truth_values, generation_correctness)
        
    return eval_dict

# ==============================================================================
# 3. 主执行流程
# ==============================================================================

def main():
    # --- 配置 ---
    TRAIN_TEST_FILE_PATH = '/home/qluai/zjs/TruthTorchLM-main/final_simplified_results_1.json'
    CALIBRATION_FILE_PATH = '/home/qluai/zjs/TruthTorchLM-main/final_simplified_results_for_calibrating.json'
    MLP_HIDDEN_LAYERS = (32, 32)
    METRICS_TO_CALCULATE = ["prr", "auroc", "auprc", "auarc"]

    # --- 步骤 1: 加载主数据集并分割 ---
    print("正在加载主数据集（用于训练和测试）...")
    with open(TRAIN_TEST_FILE_PATH, 'r', encoding='utf-8') as f:
        all_data = json.load(f)
    X = np.array([item['normalized_truth_values'] for item in all_data])
    y = np.array([item['label'] for item in all_data])
    
    # 新增：创建原始索引数组
    indices = np.arange(len(X))

    # 新增：在分割数据时，一并分割索引数组
    X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
        X, y, indices, test_size=0.3, random_state=42, stratify=y
    )
    print(f"主数据集已分割为 {len(X_train)} 训练样本和 {len(X_test)} 测试样本。")

    # 新增：将测试集对应的原始索引保存到文件
    with open("test_indices_2.json", "w", encoding='utf-8') as f:
        json.dump(idx_test.tolist(), f)
    print("测试集对应的原始索引已保存至 'test_indices.json_2'\n")


    # --- 步骤 2: 训练模型 ---
    print("--- 正在训练 LR 和 MLP 模型 ---")
    lr_model = LogisticRegression(max_iter=1000, random_state=42).fit(X_train, y_train)
    mlp_model = MLPClassifier(hidden_layer_sizes=MLP_HIDDEN_LAYERS, max_iter=1000, random_state=42).fit(X_train, y_train)
    print("模型训练完成。\n")

    # --- 步骤 3: 加载校准数据集并拟合校准器 ---
    print("--- 正在加载校准数据集并拟合校准器 ---")
    with open(CALIBRATION_FILE_PATH, 'r', encoding='utf-8') as f:
        calib_data = json.load(f)
    X_calib = np.array([item['normalized_truth_values'] for item in calib_data])
    y_calib = np.array([item['label'] for item in calib_data])

    lr_calibrator = IsotonicRegression()
    lr_uncalibrated_scores = lr_model.predict_proba(X_calib)[:, 1]
    lr_calibrator.fit(y_calib, lr_uncalibrated_scores)

    mlp_calibrator = IsotonicRegression()
    mlp_uncalibrated_scores = mlp_model.predict_proba(X_calib)[:, 1]
    mlp_calibrator.fit(y_calib, mlp_uncalibrated_scores)
    print("校准器拟合完成。\n")

    # --- 步骤 4: 在测试集上评估 ---
    print("--- 正在测试集上评估模型性能 ---")
    truth_values_lr_uncalib = lr_model.predict_proba(X_test)[:, 1]
    truth_values_mlp_uncalib = mlp_model.predict_proba(X_test)[:, 1]

    truth_values_lr_calib = np.array([lr_calibrator(score) for score in truth_values_lr_uncalib])
    truth_values_mlp_calib = np.array([mlp_calibrator(score) for score in truth_values_mlp_uncalib])

    lr_metrics_uncalib = metric_score(METRICS_TO_CALCULATE, y_test, truth_values_lr_uncalib)
    lr_metrics_calib = metric_score(METRICS_TO_CALCULATE, y_test, truth_values_lr_calib)
    mlp_metrics_uncalib = metric_score(METRICS_TO_CALCULATE, y_test, truth_values_mlp_uncalib)
    mlp_metrics_calib = metric_score(METRICS_TO_CALCULATE, y_test, truth_values_mlp_calib)

    # --- 步骤 5: 展示结果 ---
    print("\n" + "="*70)
    print(" " * 25 + "评估结果对比")
    print("="*70)
    print(f"{'指标':<10} | {'LR (未校准)':<15} | {'LR (已校准)':<15} | {'MLP (未校准)':<15} | {'MLP (已校准)':<15}")
    print("-" * 70)
    for metric in METRICS_TO_CALCULATE:
        lru = lr_metrics_uncalib.get(metric, 'N/A')
        lrc = lr_metrics_calib.get(metric, 'N/A')
        mlpu = mlp_metrics_uncalib.get(metric, 'N/A')
        mlpc = mlp_metrics_calib.get(metric, 'N/A')
        print(f"{metric.upper():<10} | {lru:<15.4f} | {lrc:<15.4f} | {mlpu:<15.4f} | {mlpc:<15.4f}")
    print("="*70)

if __name__ == "__main__":
    main()
```

```python
# -*- coding: utf-8 -*-
"""
集成等渗回归校准功能，对 LR 和多种 MLP 模型生成的真值分数进行校准和评估，
并可视化不同隐藏层结构对指标的影响（英文标注版）。

步骤：
1. 加载主数据集，并分割为训练集和测试集，同时保存测试集原始索引。
2. 加载独立的校准数据集。
3. 训练对照模型 LR，以及若干不同结构的 MLP。
4. 为每个模型拟合 IsotonicRegression 校准器。
5. 在测试集上生成未校准和校准后的分数，并计算四项指标。
6. 汇总所有模型结果到 DataFrame，打印并可视化（英文标签）。

⚠️ 请根据实际文件路径调整 TRAIN_TEST_FILE_PATH 与 CALIBRATION_FILE_PATH。
"""

# 1. Imports ────────────────────────────────────────────────────────────────
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_auc_score, precision_recall_curve, auc
import sklearn.isotonic as isotonic

# ============================================================================== 
# 2. 校准器与评估函数
# ============================================================================== 
class Normalizer:
    def fit(self, *args, **kwargs):
        pass
    def __call__(self, *args, **kwargs):
        pass

class IsotonicRegression(Normalizer):
    """sklearn.isotonic.IsotonicRegression 的封装，简化调用"""
    def __init__(self, y_min: float = 0.0, y_max: float = 1.0):
        self.iso_reg = isotonic.IsotonicRegression(
            y_min=y_min, y_max=y_max, out_of_bounds="clip"
        )
    def fit(self, generation_performance_scores, truth_values):
        self.iso_reg.fit(truth_values, generation_performance_scores)
    def __call__(self, truth_value):
        return self.iso_reg.predict([truth_value])[0]

def area_under_accuracy_coverage_curve(t_s, acc):
    df = pd.DataFrame({"t_s": t_s, "acc": acc}).sort_values("t_s", ascending=False)
    df["acc_mean"] = df["acc"].expanding().mean()
    return auc(np.linspace(0, 1, len(df)), df["acc_mean"])

def normalize(target):
    """简单按 min‑max 归一化到 [0,1]"""
    min_t, max_t = np.min(target), np.max(target)
    if np.isclose(min_t, max_t):
        min_t -= 1
        max_t += 1
    return (np.array(target) - min_t) / (max_t - min_t)

def prediction_rejection_curve(estimator, target):
    """PRR 曲线下的面积（越大越好）"""
    target = normalize(target)
    ue = np.array(estimator)
    num_obs = len(ue)
    ue_argsort = np.argsort(ue)[::-1]
    sorted_metrics = np.array(target)[ue_argsort]
    cumsum = np.cumsum(sorted_metrics)
    scores = (cumsum / np.arange(1, num_obs + 1))
    return np.sum(scores) / num_obs

def get_random_scores(function, metrics, num_iter=1000, seed=42):
    np.random.seed(seed)
    rand_scores_template = np.arange(len(metrics))
    value = []
    for _ in range(num_iter):
        np.random.shuffle(rand_scores_template)
        rand_val = function(rand_scores_template, metrics)
        value.append(rand_val)
    return np.mean(value)

def metric_score(metric_names, generation_correctness, truth_values, seed=0):
    """计算四项指标：PRR、AUROC、AUPRC、AUARC"""
    eval_dict = {}
    generation_correctness = np.array(generation_correctness)
    truth_values = np.array(truth_values)
    
    if "prr" in metric_names:
        ue_prr = prediction_rejection_curve(truth_values, generation_correctness)
        orc_prr = prediction_rejection_curve(generation_correctness, generation_correctness)
        rand_prr = get_random_scores(prediction_rejection_curve, generation_correctness, seed=seed)
        eval_dict["prr"] = 0.0 if np.isclose(orc_prr, rand_prr) else (ue_prr - rand_prr) / (orc_prr - rand_prr)
            
    if "auroc" in metric_names:
        try:
            eval_dict["auroc"] = roc_auc_score(generation_correctness, truth_values)
        except ValueError:
            eval_dict["auroc"] = 0.5  # 若标签只有单类
    
    if "auprc" in metric_names:
        precision, recall, _ = precision_recall_curve(generation_correctness, truth_values)
        eval_dict["auprc"] = auc(recall, precision)
        
    if "auarc" in metric_names:
        eval_dict["auarc"] = area_under_accuracy_coverage_curve(truth_values, generation_correctness)
        
    return eval_dict

# ============================================================================== 
# 3. 主执行流程
# ============================================================================== 
def main():
    # --- 配置 --------------------------------------------------------------
    TRAIN_TEST_FILE_PATH = '/home/qluai/zjs/TruthTorchLM-main/final_simplified_results_1.json'
    CALIBRATION_FILE_PATH = '/home/qluai/zjs/TruthTorchLM-main/final_simplified_results_for_calibrating.json'
    
    # 需要遍历的 MLP 结构：元组表示每层神经元数
    MLP_SPACE = [
        (32,),           # 1 层 32
        (64,),
        (32, 32),        # 2 层 32‑32
        (64, 32),
        (128, 64, 32)    # 3 层 128‑64‑32
    ]
    METRICS_TO_CALCULATE = ["prr", "auroc", "auprc", "auarc"]

    # --- 步骤 1: 加载主数据集并分割 ---------------------------------------
    print("Loading main dataset…")
    with open(TRAIN_TEST_FILE_PATH, 'r', encoding='utf-8') as f:
        all_data = json.load(f)
    X = np.array([item['normalized_truth_values'] for item in all_data])
    y = np.array([item['label'] for item in all_data])
    indices = np.arange(len(X))
    
    X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
        X, y, indices, test_size=0.3, random_state=42, stratify=y
    )
    print(f"Split done: train {len(X_train)}, test {len(X_test)}")
    
    # 保存测试集原始索引
    with open("test_indices_3.json", "w", encoding='utf-8') as f:
        json.dump(idx_test.tolist(), f)
    print("Test indices saved to test_indices.json\n")

    # --- 步骤 2: 训练 LR（基线） -----------------------------------------
    print("Training baseline Logistic Regression …")
    lr_model = LogisticRegression(max_iter=1000, random_state=42).fit(X_train, y_train)
    
    # --- 步骤 3: 加载校准数据集并为 LR 拟合校准器 -------------------------
    with open(CALIBRATION_FILE_PATH, 'r', encoding='utf-8') as f:
        calib_data = json.load(f)
    X_calib = np.array([item['normalized_truth_values'] for item in calib_data])
    y_calib = np.array([item['label'] for item in calib_data])

    lr_calibrator = IsotonicRegression()
    lr_uncalib_scores_calib = lr_model.predict_proba(X_calib)[:, 1]
    lr_calibrator.fit(y_calib, lr_uncalib_scores_calib)

    # --- 步骤 4: 遍历 MLP 结构 -------------------------------------------
    results = []
    for hid in MLP_SPACE:
        print(f"\nTraining MLP{hid} …")
        mlp = MLPClassifier(hidden_layer_sizes=hid, max_iter=1000, random_state=42).fit(X_train, y_train)

        # 为该 MLP 拟合校准器
        mlp_cal = IsotonicRegression()
        mlp_uncalib_scores_calib = mlp.predict_proba(X_calib)[:, 1]
        mlp_cal.fit(y_calib, mlp_uncalib_scores_calib)

        # 测试集预测
        mlp_uncalib_test = mlp.predict_proba(X_test)[:, 1]
        mlp_calib_test = np.array([mlp_cal(s) for s in mlp_uncalib_test])

        # 计算指标
        met_uncal = metric_score(METRICS_TO_CALCULATE, y_test, mlp_uncalib_test)
        met_cal = metric_score(METRICS_TO_CALCULATE, y_test, mlp_calib_test)

        # 收集结果
        results.append({
            "arch": "×".join(map(str, hid)),
            **{f"{m}_uncal": met_uncal[m] for m in METRICS_TO_CALCULATE},
            **{f"{m}_cal":   met_cal[m]   for m in METRICS_TO_CALCULATE},
        })

    # --- 步骤 5: 把 LR 指标也加入 ----------------------------------------
    lr_uncalib_test = lr_model.predict_proba(X_test)[:, 1]
    lr_calib_test = np.array([lr_calibrator(s) for s in lr_uncalib_test])
    lr_uncal_metrics = metric_score(METRICS_TO_CALCULATE, y_test, lr_uncalib_test)
    lr_cal_metrics = metric_score(METRICS_TO_CALCULATE, y_test, lr_calib_test)
    results.append({
        "arch": "LR",
        **{f"{m}_uncal": lr_uncal_metrics[m] for m in METRICS_TO_CALCULATE},
        **{f"{m}_cal":   lr_cal_metrics[m]   for m in METRICS_TO_CALCULATE},
    })

    # --- 步骤 6: 输出并可视化 -------------------------------------------
    df = pd.DataFrame(results).sort_values("arch")
    print("\n================ Evaluation Results ================")
    print(df.round(4).to_string(index=False))
    
    # 可选：保存 CSV
    # df.to_csv("mlp_results.csv", index=False)

    # === Visualization (English labels) ===================================
    ### ENGLISH ###
    plt.figure(figsize=(10, 6))
    for metric in ["auroc", "auprc"]:       # add more if needed
        plt.plot(df["arch"], df[f"{metric}_uncal"], marker="o",
                 label=f"{metric.upper()} Uncalibrated")
        plt.plot(df["arch"], df[f"{metric}_cal"], marker="s", linestyle="--",
                 label=f"{metric.upper()} Calibrated")
    plt.ylabel("Score")
    plt.xlabel("Model architecture (hidden layers)")
    plt.title("Effect of MLP architecture on metrics")
    plt.legend()
    plt.grid(alpha=.3)
    plt.tight_layout()
    plt.show()

# ---------------------------------------------------------------------------
if __name__ == "__main__":
    main()
```

```python
# -*- coding: utf-8 -*-
"""
集成等渗回归校准功能，对 LR 和多种 MLP 模型生成的真值分数进行校准和评估，
并可视化不同隐藏层结构对指标的影响（英文标注版）。

步骤：
1. 加载主数据集，并分割为训练集和测试集，同时保存测试集原始索引。
2. 加载独立的校准数据集。
3. 训练对照模型 LR，以及若干不同结构的 MLP。
4. 为每个模型拟合 IsotonicRegression 校准器。
5. 在测试集上生成未校准和校准后的分数，并计算四项指标。
6. 汇总所有模型结果到 DataFrame，打印并可视化（英文标签）。

⚠️ 请根据实际文件路径调整 TRAIN_TEST_FILE_PATH 与 CALIBRATION_FILE_PATH。
"""

# 1. Imports ────────────────────────────────────────────────────────────────
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_auc_score, precision_recall_curve, auc
import sklearn.isotonic as isotonic

# ==============================================================================
# 2. 校准器与评估函数
# ==============================================================================
class Normalizer:
    def fit(self, *args, **kwargs):
        pass
    def __call__(self, *args, **kwargs):
        pass

class IsotonicRegression(Normalizer):
    """sklearn.isotonic.IsotonicRegression 的封装，简化调用"""
    def __init__(self, y_min: float = 0.0, y_max: float = 1.0):
        self.iso_reg = isotonic.IsotonicRegression(
            y_min=y_min, y_max=y_max, out_of_bounds="clip"
        )
    def fit(self, generation_performance_scores, truth_values):
        self.iso_reg.fit(truth_values, generation_performance_scores)
    def __call__(self, truth_value):
        return self.iso_reg.predict([truth_value])[0]

def area_under_accuracy_coverage_curve(t_s, acc):
    df = pd.DataFrame({"t_s": t_s, "acc": acc}).sort_values("t_s", ascending=False)
    df["acc_mean"] = df["acc"].expanding().mean()
    return auc(np.linspace(0, 1, len(df)), df["acc_mean"])

def normalize(target):
    """简单按 min‑max 归一化到 [0,1]"""
    min_t, max_t = np.min(target), np.max(target)
    if np.isclose(min_t, max_t):
        min_t -= 1
        max_t += 1
    return (np.array(target) - min_t) / (max_t - min_t)

def prediction_rejection_curve(estimator, target):
    """PRR 曲线下的面积（越大越好）"""
    target = normalize(target)
    ue = np.array(estimator)
    num_obs = len(ue)
    ue_argsort = np.argsort(ue)[::-1]
    sorted_metrics = np.array(target)[ue_argsort]
    cumsum = np.cumsum(sorted_metrics)
    scores = (cumsum / np.arange(1, num_obs + 1))
    return np.sum(scores) / num_obs

def get_random_scores(function, metrics, num_iter=1000, seed=42):
    np.random.seed(seed)
    rand_scores_template = np.arange(len(metrics))
    value = []
    for _ in range(num_iter):
        np.random.shuffle(rand_scores_template)
        rand_val = function(rand_scores_template, metrics)
        value.append(rand_val)
    return np.mean(value)

def metric_score(metric_names, generation_correctness, truth_values, seed=0):
    """计算四项指标：PRR、AUROC、AUPRC、AUARC"""
    eval_dict = {}
    generation_correctness = np.array(generation_correctness)
    truth_values = np.array(truth_values)

    if "prr" in metric_names:
        ue_prr = prediction_rejection_curve(truth_values, generation_correctness)
        orc_prr = prediction_rejection_curve(generation_correctness, generation_correctness)
        rand_prr = get_random_scores(prediction_rejection_curve, generation_correctness, seed=seed)
        eval_dict["prr"] = 0.0 if np.isclose(orc_prr, rand_prr) else (ue_prr - rand_prr) / (orc_prr - rand_prr)

    if "auroc" in metric_names:
        try:
            eval_dict["auroc"] = roc_auc_score(generation_correctness, truth_values)
        except ValueError:
            eval_dict["auroc"] = 0.5  # 若标签只有单类

    if "auprc" in metric_names:
        precision, recall, _ = precision_recall_curve(generation_correctness, truth_values)
        eval_dict["auprc"] = auc(recall, precision)

    if "auarc" in metric_names:
        eval_dict["auarc"] = area_under_accuracy_coverage_curve(truth_values, generation_correctness)

    return eval_dict

# ==============================================================================
# 3. 主执行流程
# ==============================================================================
def main():
    # --- 配置 --------------------------------------------------------------
    # ⚠️ 请确保这里的路径是正确的
    TRAIN_TEST_FILE_PATH = '/home/qluai/zjs/TruthTorchLM-main/final_simplified_results_1.json'
    CALIBRATION_FILE_PATH = '/home/qluai/zjs/TruthTorchLM-main/final_simplified_results_for_calibrating.json'

    # 需要遍历的 MLP 结构：元组表示每层神经元数
    MLP_SPACE = [
        (32,),          # 1 层 32
        (64,),
        (32, 32),       # 2 层 32‑32
        (64, 32),
        (128, 64, 32)   # 3 层 128‑64‑32
    ]
    METRICS_TO_CALCULATE = ["prr", "auroc", "auprc", "auarc"]

    # --- 步骤 1: 加载主数据集并分割 ---------------------------------------
    print("Loading main dataset…")
    with open(TRAIN_TEST_FILE_PATH, 'r', encoding='utf-8') as f:
        all_data = json.load(f)
    X = np.array([item['normalized_truth_values'] for item in all_data])
    y = np.array([item['label'] for item in all_data])
    indices = np.arange(len(X))

    X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
        X, y, indices, test_size=0.3, random_state=42, stratify=y
    )
    print(f"Split done: train {len(X_train)}, test {len(X_test)}")

    # 保存测试集原始索引
    with open("test_indices_3.json", "w", encoding='utf-8') as f:
        json.dump(idx_test.tolist(), f)
    print("Test indices saved to test_indices.json\n")

    # --- 步骤 2: 训练 LR（基线） -----------------------------------------
    print("Training baseline Logistic Regression …")
    lr_model = LogisticRegression(max_iter=1000, random_state=42).fit(X_train, y_train)

    # --- 步骤 3: 加载校准数据集并为 LR 拟合校准器 -------------------------
    with open(CALIBRATION_FILE_PATH, 'r', encoding='utf-8') as f:
        calib_data = json.load(f)
    X_calib = np.array([item['normalized_truth_values'] for item in calib_data])
    y_calib = np.array([item['label'] for item in calib_data])

    lr_calibrator = IsotonicRegression()
    lr_uncalib_scores_calib = lr_model.predict_proba(X_calib)[:, 1]
    lr_calibrator.fit(y_calib, lr_uncalib_scores_calib)

    # --- 步骤 4: 遍历 MLP 结构 -------------------------------------------
    results = []
    for hid in MLP_SPACE:
        print(f"\nTraining MLP{hid} …")
        mlp = MLPClassifier(hidden_layer_sizes=hid, max_iter=1000, random_state=42).fit(X_train, y_train)

        # 为该 MLP 拟合校准器
        mlp_cal = IsotonicRegression()
        mlp_uncalib_scores_calib = mlp.predict_proba(X_calib)[:, 1]
        mlp_cal.fit(y_calib, mlp_uncalib_scores_calib)

        # 测试集预测
        mlp_uncalib_test = mlp.predict_proba(X_test)[:, 1]
        mlp_calib_test = np.array([mlp_cal(s) for s in mlp_uncalib_test])

        # 计算指标
        met_uncal = metric_score(METRICS_TO_CALCULATE, y_test, mlp_uncalib_test)
        met_cal = metric_score(METRICS_TO_CALCULATE, y_test, mlp_calib_test)

        # 收集结果
        results.append({
            "arch": "×".join(map(str, hid)),
            **{f"{m}_uncal": met_uncal[m] for m in METRICS_TO_CALCULATE},
            **{f"{m}_cal":   met_cal[m]   for m in METRICS_TO_CALCULATE},
        })

    # --- 步骤 5: 把 LR 指标也加入 ----------------------------------------
    lr_uncalib_test = lr_model.predict_proba(X_test)[:, 1]
    lr_calib_test = np.array([lr_calibrator(s) for s in lr_uncalib_test])
    lr_uncal_metrics = metric_score(METRICS_TO_CALCULATE, y_test, lr_uncalib_test)
    lr_cal_metrics = metric_score(METRICS_TO_CALCULATE, y_test, lr_calib_test)
    results.append({
        "arch": "LR",
        **{f"{m}_uncal": lr_uncal_metrics[m] for m in METRICS_TO_CALCULATE},
        **{f"{m}_cal":   lr_cal_metrics[m]   for m in METRICS_TO_CALCULATE},
    })

    # --- 步骤 6: 输出并可视化 -------------------------------------------
    df = pd.DataFrame(results)
    
    # --- MODIFICATION START ---
    # 定义绘图顺序，将 LR 放在最前面作为基线进行比较
    plot_order = ['LR'] + ["×".join(map(str, hid)) for hid in MLP_SPACE]
    # 将 'arch' 列转为有序的分类类型
    df['arch'] = pd.Categorical(df['arch'], categories=plot_order, ordered=True)
    # 按自定义顺序排序
    df = df.sort_values('arch')
    # --- MODIFICATION END ---
    
    print("\n================ Evaluation Results ================")
    print(df.round(4).to_string(index=False))

    # 可选：保存 CSV
    # df.to_csv("mlp_lr_results.csv", index=False)

    # === Visualization (English labels) ===================================
    plt.figure(figsize=(12, 7))
    
    # 选择要可视化的指标
    metrics_to_plot = ["auroc", "auprc", "prr", "auarc"] 
    
    # 为每个指标创建一个子图
    for i, metric in enumerate(metrics_to_plot):
        ax = plt.subplot(2, 2, i + 1)
        ax.plot(df["arch"], df[f"{metric}_uncal"], marker="o", linestyle="-", label="Uncalibrated")
        ax.plot(df["arch"], df[f"{metric}_cal"], marker="s", linestyle="--", label="Calibrated")
        ax.set_title(f"{metric.upper()} Score")
        ax.set_ylabel("Score")
        ax.grid(alpha=.4)
        ax.legend()
        # 旋转X轴标签以防重叠
        ax.tick_params(axis='x', rotation=45)

    # 全局标题和X轴标签
    plt.suptitle("Model Performance Comparison (Before and After Isotonic Calibration)", fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # 调整布局为全局标题留出空间
    plt.show()


# ---------------------------------------------------------------------------
if __name__ == "__main__":
    main()
```

```python
# -*- coding: utf-8 -*-
"""
Compare LR, multiple MLP architectures, and additional uncertainty baselines
with Isotonic calibration.  Visualize AUROC & PRR in English, highlighting
the best model on each metric.
"""

# ───────────────────────── Imports ────────────────────────────────────────
import json, math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_auc_score, precision_recall_curve, auc
import sklearn.isotonic as isotonic

# ───────────────────────── Helper classes / metrics ──────────────────────
class IsotonicRegression:
    def __init__(self, y_min=0.0, y_max=1.0):
        self.iso = isotonic.IsotonicRegression(y_min=y_min, y_max=y_max,
                                               out_of_bounds="clip")
    def fit(self, scores, labels):
        self.iso.fit(labels, scores)
    def __call__(self, s):
        return self.iso.predict([s])[0]

def normalize(arr):
    a = np.asarray(arr)
    lo, hi = a.min(), a.max()
    if math.isclose(lo, hi):
        hi += 1
    return (a - lo) / (hi - lo)

def prediction_rejection_curve(estimator, target):
    target = normalize(target)
    ue = np.asarray(estimator)
    order = np.argsort(ue)[::-1]
    cum = np.cumsum(target[order])
    return (cum / np.arange(1, len(ue) + 1)).mean()

def get_random_scores(func, metrics, num_iter=1000, seed=42):
    rng = np.random.default_rng(seed)
    idx = np.arange(len(metrics))
    vals = []
    for _ in range(num_iter):
        rng.shuffle(idx)
        vals.append(func(idx, metrics))
    return np.mean(vals)

def metric_score(metric_names, y_true, y_score, seed=0):
    out = {}
    y_true = np.asarray(y_true)
    y_score = np.asarray(y_score)

    if "auroc" in metric_names:
        try:
            out["auroc"] = roc_auc_score(y_true, y_score)
        except ValueError:
            out["auroc"] = 0.5

    if "prr" in metric_names:
        ue_prr = prediction_rejection_curve(y_score, y_true)
        orc_prr = prediction_rejection_curve(y_true, y_true)
        rnd_prr = get_random_scores(prediction_rejection_curve, y_true, seed=seed)
        out["prr"] = 0.0 if math.isclose(orc_prr, rnd_prr) else (ue_prr - rnd_prr) / (orc_prr - rnd_prr)

    return out

# ───────────────────────────── Main pipeline ─────────────────────────────
def main():
    # ========== Config ====================================================
    TRAIN_TEST_FILE_PATH = "/home/qluai/zjs/TruthTorchLM-main/final_simplified_results_1.json"
    CALIBRATION_FILE_PATH = "/home/qluai/zjs/TruthTorchLM-main/final_simplified_results_for_calibrating.json"

    MLP_SPACE = [(32,), (64,), (32, 32), (64, 32), (128, 64, 32)]
    METRICS_TO_CALCULATE = ["auroc", "prr"]

    # ---------- External baselines ---------------------------------------
    BASELINES = {
        "Entropy": {"auroc": 0.6927162142115413, "prr": 0.4607823926513025},
        "SemanticEntropy": {"auroc": 0.6904020175048213, "prr": 0.4607163100091704},
        "MARS": {"auroc": 0.735790931118034, "prr": 0.5649332870679039},
        "LARS": {"auroc": 0.5977418450938701, "prr": 0.21846993984421925},
        "SentSAR": {"auroc": 0.686169213272017, "prr": 0.45369971284616945},
        "TokenSAR": {"auroc": 0.7190723433714088, "prr": 0.504908381574678},
        "Confidence": {"auroc": 0.7159735614564274, "prr": 0.49311507496600443},
        "Inside": {"auroc": 0.6454119896487498, "prr": 0.27650469586337206},
        "AttentionScore": {"auroc": 0.4968352865549127, "prr": 0.008750928480825409},
        "EccUncertainty": {"auroc": 0.6617234501969704, "prr": 0.261847846937212},
        "EccConfidence": {"auroc": 0.6519622871647794, "prr": 0.25998183231230926},
        "MxDegConfidence": {"auroc": 0.6210947930574099, "prr": 0.21682945396122594},
        "MxDegUncertainty": {"auroc": 0.6272396116632877, "prr": 0.22910378924642968},
        "SumEigenUncertainty": {"auroc": 0.6273385089584467, "prr": 0.229175530009427},
        "PTrue": {"auroc": 0.7291747020718984, "prr": 0.4903621547957103},
        "VerbalizedConf": {"auroc": 0.5478547528391766, "prr": 0.09764097900736275},
        "SelfDetection": {"auroc": 0.517890520694259, "prr": -0.01979791431245392},
        "DirEntailGraph": {"auroc": 0.6581928167597949, "prr": 0.23046720082983185},
    }

    # ========== Load & split data ========================================
    with open(TRAIN_TEST_FILE_PATH, "r", encoding="utf-8") as f:
        data = json.load(f)
    X = np.array([it["normalized_truth_values"] for it in data])
    y = np.array([it["label"] for it in data])

    X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

    # ========== Calibration set ==========================================
    with open(CALIBRATION_FILE_PATH, "r", encoding="utf-8") as f:
        calib = json.load(f)
    X_c = np.array([it["normalized_truth_values"] for it in calib])
    y_c = np.array([it["label"] for it in calib])

    # ========== Baseline LR =============================================
    lr = LogisticRegression(max_iter=1000, random_state=42).fit(X_tr, y_tr)
    lr_cal = IsotonicRegression()
    lr_cal.fit(lr.predict_proba(X_c)[:, 1], y_c)

    results = []

    # LR metrics (uncalibrated & calibrated)
    lr_unc = lr.predict_proba(X_te)[:, 1]
    lr_cal_scores = np.array([lr_cal(s) for s in lr_unc])
    results.append({"model": "LR‑Uncal", **metric_score(METRICS_TO_CALCULATE, y_te, lr_unc)})
    results.append({"model": "LR‑Cal", **metric_score(METRICS_TO_CALCULATE, y_te, lr_cal_scores)})

    # ========== Iterate MLP architectures ===============================
    for hid in MLP_SPACE:
        mlp = MLPClassifier(hidden_layer_sizes=hid, max_iter=1000, random_state=42).fit(X_tr, y_tr)

        cal = IsotonicRegression()
        cal.fit(mlp.predict_proba(X_c)[:, 1], y_c)

        unc = mlp.predict_proba(X_te)[:, 1]
        cal_sc = np.array([cal(s) for s in unc])

        name = "MLP" + "×".join(map(str, hid))
        results.append({"model": f"{name}‑Uncal", **metric_score(METRICS_TO_CALCULATE, y_te, unc)})
        results.append({"model": f"{name}‑Cal", **metric_score(METRICS_TO_CALCULATE, y_te, cal_sc)})

    # ========== Add external baselines ==================================
    for name, vals in BASELINES.items():
        results.append({"model": name, "auroc": vals["auroc"], "prr": vals["prr"]})

    # ========== DataFrame & sort / reset index ===========================
    df = pd.DataFrame(results)
    df["sortkey"] = df["model"].apply(
        lambda x: (0 if x.startswith("LR") else 1 if x.startswith("MLP") else 2, x)
    )
    df.sort_values("sortkey", inplace=True)
    df.drop(columns="sortkey", inplace=True)
    df.reset_index(drop=True, inplace=True)          # ★★ 关键修复：重置索引 ★★

    print("\n========= Summary (AUROC / PRR) =========")
    print(df.round(4).to_string(index=False))

    # ========== Visualization ===========================================
    fig, axes = plt.subplots(2, 1, figsize=(12, 8), sharex=True, gridspec_kw={"hspace": 0.25})
    x = np.arange(len(df))

    # 1) find best indices (now aligned with x)
    best_auroc_idx = df["auroc"].idxmax()
    best_prr_idx = df["prr"].idxmax()

    print(f"\n👑  Best AUROC : {df.loc[best_auroc_idx, 'model']} "
          f"({df.loc[best_auroc_idx, 'auroc']:.4f})")
    print(f"👑  Best PRR   : {df.loc[best_prr_idx, 'model']} "
          f"({df.loc[best_prr_idx, 'prr']:.4f})")

    # 2) AUROC plot
    axes[0].plot(x, df["auroc"], marker="o")
    axes[0].scatter(best_auroc_idx, df.loc[best_auroc_idx, "auroc"],
                    marker="*", s=180, color="red", zorder=5)
    axes[0].annotate(f"{df.loc[best_auroc_idx, 'auroc']:.3f}",
                     xy=(best_auroc_idx, df.loc[best_auroc_idx, 'auroc']),
                     xytext=(0, 8), textcoords="offset points",
                     ha="center", color="red", fontsize=8)
    axes[0].set_ylabel("AUROC")
    axes[0].set_title("AUROC across models")
    axes[0].grid(alpha=.3)

    # 3) PRR plot
    axes[1].plot(x, df["prr"], marker="s", color="tab:orange")
    axes[1].scatter(best_prr_idx, df.loc[best_prr_idx, "prr"],
                    marker="*", s=180, color="red", zorder=5)
    axes[1].annotate(f"{df.loc[best_prr_idx, 'prr']:.3f}",
                     xy=(best_prr_idx, df.loc[best_prr_idx, 'prr']),
                     xytext=(0, 8), textcoords="offset points",
                     ha="center", color="red", fontsize=8)
    axes[1].set_ylabel("PRR")
    axes[1].set_title("PRR across models")
    axes[1].grid(alpha=.3)

    # 4) x‑tick labels
    axes[1].set_xticks(x)
    axes[1].set_xticklabels(df["model"], rotation=90, fontsize=8)

    plt.tight_layout()
    plt.show()


# ───────────────────────────────────────────────────────────────────────────
if __name__ == "__main__":
    main()
```
